## Cities on a square

The aim of this exercise is to find the shortest path that connects 34 cities. The constraints are that

1 - you can't visit a city more than once

2 - you have to go back to the city that you started from

This problem naturally has a degeneracy linked to the fact that the best path could be made starting from any city. It's as if we had a $U(1)$ symmetry: it is a good idea to break it **explicitly** in order to reduce the space in which we look for the solution. This entails conventionally taking one city to be the starting point for all the paths.\
It is to say that, though, that we still have a $Z_2$ residual symmetry, because the path can be walked clockwise or anti-clockwise

In the code the locations of the cities are randomly generated in a square.

The study has been made with a population of 3000 individuals, running 200 generations

```python
Ngens = 200
Plot(Ngens, "Square")
```

We can see that just about after 80 generations the solution stabilizes itself on what we take to be our final solution. Performing the simulation multiple times (changing the seed), we have observed that 4.98426 seems to be the lowest achievable length by this algorithm.

From the plot of the lengths we can see that even after the best individual is created the genetic processes do not stop, and the population keeps improving as if its genes keep spreading in the population's gene pool. The trend for the best lenght presents no bumps, while the average over the best half reflects the behavior of a genetic algorithm: solutions that are far from optimal are created and still contribute to the genetic heritage of what is going to be the best individual
