# Parallel Computing - Travelling Salesman Problem

```python
from array import array
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from matplotlib.pyplot import figure
from PlotFunctions import read, PlotLengths, PlotPath
```

```python
N_cores = 8
```

The algorithm is run on 8 cores with 300 individuals, first with an independent search, and then with migrations.

A migration consists in the fact that, every 25 generations, a random sequence dictates the exchange of the best individual among the populations hosted on the different cores.

We are going to show how parallel computing gives a huge boost to the search of the optimal solution: with the **same number of individuals** and running the simulation for the **same number of generations**, the search with migrations cuts down the length of the path by a significative margin

## Independent Search

```python
PlotLengths('Independent', N_cores)
```

The width of the branch in the graph shows how each core stays on its own track during the search, with results that can vary a lot among themselves. The shortest length among the various performances is reported above and the corresponding path is plotted below

```python
PlotPath('Independent')
```

From this we can see that the blue lines are quite dense and a lot of crossings are present 

## Search with Migrations

```python
PlotLengths('Migrations', N_cores)
```

Here we can see the migrations in action in the fact that about after generation 600 the best individuals tend to converge towards being a single branch

```python
PlotPath('Migrations')
```
